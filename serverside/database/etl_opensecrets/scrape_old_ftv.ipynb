```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
```

```python
options = webdriver.ChromeOptions()
options.add_argument("ignore-certificate-errors")
driver = webdriver.Chrome(options=options)
```

```python
import pandas as pd
import time
```

```python
crp_ids = pd.read_csv("just_crp_ids.csv", header=None)
jsons = []
for crp in crp_ids[0]:
    print(crp)
    driver.get(f"https://api.fundstovotes.info/topten?crp_id={crp}")
    time.sleep(1)
    jsons.append(driver.find_element(By.TAG_NAME, "body").text)
```

```python
jsons[0]
```

```python
import json
```

```python
flattened_json = {}
flattened_json["cand_name"] = []
flattened_json["cid"] = []
flattened_json["cycle"] = []
flattened_json["last_updated"] = []
for i in range(10):
    flattened_json[f'industry_code{i}'] = []
    flattened_json[f'industry_name{i}'] = []
    flattened_json[f'indivs{i}'] = []
    flattened_json[f'pacs{i}'] = []
    flattened_json[f'total{i}'] = []
```

```python
for data_point in jsons:
    loaded = json.loads(data_point)
    for i, industry in enumerate(loaded["industry"]):
        flattened_json[f'industry_code{i}'].append(industry["industry_code"])
        flattened_json[f'industry_name{i}'].append(industry["industry_name"])
        flattened_json[f'indivs{i}'].append(industry["indivs"])
        flattened_json[f'pacs{i}'].append(industry["pacs"])
        flattened_json[f'total{i}'].append(industry["total"])
        
    attributes = loaded["@attributes"]
    flattened_json["cand_name"].append(attributes["cand_name"])
    flattened_json["cid"].append(attributes["cid"])
    flattened_json["cycle"].append(attributes["cycle"])
    flattened_json["last_updated"].append(attributes["last_updated"])
```

```python
pd.DataFrame(flattened_json).to_csv("test.csv")
```
